```python
import pandas as pd
```

```python
file="Resources/purchase_data.csv"
purchase_data = pd.read_csv(file)
purchase_data
```

# Player Count

```python
players = len(purchase_data["SN"].unique())
players_df = pd.DataFrame([players]).rename(columns={0:"Total Players"})
players_df
```

# Purchasing Analysis

```python
item_count = len(purchase_data["Item ID"].unique())
purchase_count = purchase_data["Purchase ID"].count()
avg_price = purchase_data["Price"].mean()
total_purchase = purchase_data["Price"].sum()

purchase = pd.DataFrame([{
    "Unique Items Count" :item_count,
    "Average Price" : avg_price,
    "Number of Purchases" : purchase_count,
    "Total Revenue" : total_purchase}])
purchase
```

# Gender Demographics

```python
gender = pd.DataFrame(purchase_data[["SN","Gender"]].value_counts())
gender = gender.reset_index()
gender = gender[["SN","Gender"]]
gender
```

```python
gender_list = pd.DataFrame(gender["Gender"].value_counts())
gender_list = gender_list.rename(columns={"Gender":"Total Count"})
gender_list["Percentage of Players"] = gender_list["Total Count"]/gender_list["Total Count"].sum()*100
gender_list["Percentage of Players"] = gender_list["Percentage of Players"].map("{:.2f}%".format)
gender_list
```

# Purchasing Analysis (Gender)

```python
gender_group = purchase_data.groupby(["Gender"]).agg({
    "Item ID": "count",
    "Price":["mean","sum"]
})
gender_group.columns = ["Purchase Count", "Avg Purchase Price","Total Purchase Value"]
gender_group["Avg Total Purchase per Person"] = gender_group["Total Purchase Value"]/gender_group["Purchase Count"]
gender_group["Avg Purchase Price"] = gender_group["Avg Purchase Price"].map("${:.2f}".format)
gender_group["Total Purchase Value"] = gender_group["Total Purchase Value"].map("${:,.2f}".format)
gender_group["Avg Total Purchase per Person"] = gender_group["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_group
```

# Age Demographics

```python
age_df = pd.DataFrame(purchase_data[["SN","Age"]].value_counts())
age_df = age_df.reset_index()
age_df = age_df[["Age"]]

agebin = [0,10,15,20,25,30,35,40,9999]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_df["Age Group"] = pd.cut(age_df["Age"], agebin, labels=labels, include_lowest=True, right=False)
age_df = age_df.groupby("Age Group").count()
age_df.columns =["Age Group Count"]
age_df["Percentage of Players"] = age_df["Age Group Count"]/age_df["Age Group Count"].sum()*100
age_df["Percentage of Players"] = age_df["Percentage of Players"].map("{:.2f}%".format)
age_df
```

# Purchasing Analysis (Age)

```python
age_purchase_df = purchase_data
age_purchase_df["Age Group"] = pd.cut(purchase_data["Age"], agebin, labels=labels, include_lowest=True, right=False)
age_purchase_df = age_purchase_df.groupby("Age Group").agg({
    "Price" : ["count","mean", "sum"]
})
age_purchase_df.columns = ["Purchase Count", "Avg Purchase Price", "Total Purchase Value"]
age_purchase_df["Avg Total Purchase per Person"] = age_purchase_df["Total Purchase Value"]/age_df["Age Group Count"]
age_purchase_df["Avg Purchase Price"] = age_purchase_df["Avg Purchase Price"].map("${:.2f}".format)
age_purchase_df["Total Purchase Value"] = age_purchase_df["Total Purchase Value"].map("${:,.2f}".format)
age_purchase_df["Avg Total Purchase per Person"] = age_purchase_df["Avg Total Purchase per Person"].map("${:.2f}".format)
#age_purchase_df = age_purchase_df.groupby(["Age Group"]).agg({
#    "Purchase Value" : []
#})

age_purchase_df
```

# Top Spenders

```python
top_spender_df = purchase_data[["SN","Price"]]
top_spender_df = top_spender_df.groupby("SN").agg({
    "Price":["count","mean","sum"]
})
top_spender_df.columns = ["Purchase Count","Average Purchase Price","Total Purchase Value"]
top_spender_df = top_spender_df.sort_values(by=["Total Purchase Value"], ascending=False)
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)
top_spender_df
```

# Most Popular Items

```python
pop_item_df = purchase_data[["Item ID","Item Name","Price"]]
pop_item_df = pop_item_df.groupby(["Item ID","Item Name"]).agg({
    "Price": ["mean","count","sum"]
})
pop_item_df.columns = ["Item Price","Purchase Count", "Total Purchase Value"]
profit_item_df = pop_item_df
pop_item_df = pop_item_df.sort_values(by=["Purchase Count"], ascending=False)
pop_item_df["Item Price"] = pop_item_df["Item Price"].map("${:.2f}".format)
pop_item_df["Total Purchase Value"] = pop_item_df["Total Purchase Value"].map("${:.2f}".format)
pop_item_df
#check_df = pop_item_df.reset_index()
#check_df = check_df.loc[check_df["Item ID"] == 92]
#check_df
```

# Most Profitable Items

```python
profit_item_df = profit_item_df.sort_values(by=["Total Purchase Value"], ascending=False)
profit_item_df["Item Price"]=profit_item_df["Item Price"].map("${:.2f}".format)
profit_item_df["Total Purchase Value"] = profit_item_df["Total Purchase Value"].map("${:.2f}".format)
profit_item_df
```

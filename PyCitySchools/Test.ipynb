```python
import pandas as pd
```

```python
group1 = [x for x in range(1,101)]
group2 = [x for x in range(101,201)]
group3 = [x for x in range(201,301)]
group4 = [x for x in range(301,401)]
```

```python
df = pd.DataFrame({
    "Group1" : group1,
    "Group2" : group2,
    "Group3" : group3,
    "Group4" : group4
})
```

```python
bins = [0,21,41,61,81,101]
labels = ["lv1","lv2","lv3","lv4",'lv5']
df["Level"] = pd.cut(df["Group1"], bins, labels=labels, include_lowest=True)
meandf = df.groupby("Level").mean()
meandf
```

```python
bins = [0,41,81,101]
labels = ["lv1","lv2","lv3"]
df["Level2"] = pd.cut(df["Group1"], bins, labels=labels, include_lowest=True)
df
```

```python
meandf2 = df.groupby("Level2").mean()
meandf2
```
